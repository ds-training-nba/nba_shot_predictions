## Dataset Exploration

```python
from datasets import load_dataset

# Load DataSet
ds = load_dataset(
    "ds-training-nba/nba_shot_data",
    data_files={
        "full": "processed/processed_20_players.parquet"
    }
)
df = ds['full'].to_pandas()
```

## Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x=df['SHOT_MADE_FLAG'])
```

We can see more than 500k throws, 300k of them are successful.

### Throws by Player

```python
sns.countplot(
    x='PLAYER_NAME',
    hue='SHOT_MADE_FLAG',
    data=df
)

plt.xticks(rotation=90, ha='right')
plt.title('Count of shot types by Player')
plt.legend(title='Shot Made')
plt.tight_layout()
plt.show()
```

### Shot Distance distribution

```python
sns.histplot(
    data=df,
    x='SHOT_DISTANCE',
    hue='SHOT_TYPE'
)
```

## Defining variable types

### Spliting columns type to diferent categories

```python
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
boolean_cols = df.select_dtypes(include=['bool']).columns

display(numerical_cols)
display(categorical_cols)
display(boolean_cols)
```

### Check Categorical columns

```python
for col in df[categorical_cols].columns:
    print(f"{col}:")
    print(f"  unique: {df[col].unique()}")
    print()
```

### Clean up unnecessary categorical columns

```python
irrelevant_cols = [
    'PCTIMESTRING', # there is TimeRemainingInPeriod instead
    'PLAYER3_NAME',
    'PLAYER2_TEAM_ABBREVIATION',
    'PLAYER2_NAME',
    'clock',
    'shotResult', # duplicate from our Target Field
    'PLAYER1_TEAM_ABBREVIATION',
    'VTM',
    'HTM'
]

df = df.drop(irrelevant_cols, axis=1, errors='ignore')
new_categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in df[new_categorical_cols].columns:
    print(f"{col}:")
    print(f"  unique: {df[col].unique()}")
    print()
```

### Update numerical flag columns to boolean

```python
bool_cols = [
    'IS_HOME',
    'IsClutchTime',
    'IsOvertime'
]

df[bool_cols] = df[bool_cols].apply(lambda x: x.astype('bool'))
```
